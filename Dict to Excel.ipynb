### Final Code 

```python
row = OrderedDict()
```

```python
years = [i+2000 for i in range(10)]
quarters = [i%4 for i in range(10)]
days = [i for i in range(10)]
bd_ids = list(np.random.randint(20000,30000,10))
bd_names = list(np.random.randint(0,1000,10))
random_values = list(np.random.uniform(low=0,high=1000,size=10))
```

```python
print(f"years: {years},\nquarters: {quarters},\ndays: {days},\nbd_ids: {bd_ids},\nbd_names: {bd_names}")
```

```python
assert len(years) == len(quarters) == len(days) == len(bd_ids) == len(bd_names)
```

```python
row[("","","BD_NAME")] = bd_ids
row[("","","BD_ID")] = bd_names
row[("","","RANDOM")] = random_values
row[("Years","Quarters","Day")] = ""
for i, (y, q, d) in enumerate(zip(years,quarters,days)):
    row[(y,q,d)] = " "
```

```python
row
```

```python
row_df = pd.DataFrame(row,index=np.arange(10))
```

```python
row_df
```

```python
row_df.to_csv("sub3.csv",index=False)
```
